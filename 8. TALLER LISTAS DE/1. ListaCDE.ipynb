**Crear: Una lista CIRCULAR DOBLEMENTE ENLAZADA**
1. Método para insertar en:
    1. Inicio
    2. Final
2. Método para eliminar nodo
3. Método para mostrar la lista en cualquiera de los 2 sentidos (Adelante hacia atrás o viceversa) 
4. Método agregar en cualquier lugar de la lista

```python
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None
        self.anterior = None

class ListaCDE:
    def __init__(self):
        self.primero = None
        self.ultimo = None

    def vacia(self):
        if self.primero == None:
            return True
        else:
            return False

    #Método para agregar nodos al inicio
    def agregar_inicio(self, dato):
        if self.vacia():
            self.primero = self.ultimo = Nodo(dato)
        else:
            aux = Nodo(dato)
            aux.siguiente = self.primero
            self.primero.anterior = aux
            self.primero = aux
        self.__unir_nodos()

    #Método para agregar nodos al final
    def agregar_final(self, dato):
        if self.vacia():
            self.primero = self.ultimo = Nodo(dato)
        else:
            aux = self.ultimo
            self.ultimo = aux.siguiente = Nodo(dato)
            self.ultimo.anterior = aux
        self.__unir_nodos()

    #Método para eliminar nodos al inicio
    def eliminar_inicio(self):
        if self.vacia():
            print("No hay elementos para eliminar")
        elif self.primero.siguiente == None:
            self.primero = self.ultimo = None
        else:
            self.primero = self.primero.siguiente
        self.__unir_nodos()

    #Método para eliminar nodos al final
    def eliminar_final(self):
        if self.vacia():
            print("No hay elementos para eliminar")
        elif self.primero.siguiente == None:
            self.primero = self.ultimo = None
        else:
            self.ultimo = self.ultimo.anterior
            self.ultimo.siguiente = None
        self.__unir_nodos()
    
    #Método para agregar nodos en cualquier lugar de la lista
    def agregar(self, dato, referencia):
        if self.vacia():
            print("No hay elementos para agregar")
        else:
            aux = self.primero
            while aux:
                if aux.dato == referencia:
                    nuevo = Nodo(dato)
                    anterior = aux.anterior
                    anterior.siguiente = nuevo
                    nuevo.siguiente = aux
                    aux.anterior = nuevo
                    nuevo.anterior = anterior
                aux = aux.siguiente
                if aux == self.primero:
                    break
    
    #Método para eliminar nodos en cualquier lugar de la lista
    def eliminar(self, referencia):
        if self.vacia():
            print("No hay elementos para eliminar")
        else:
            aux = self.primero
            while aux:
                if aux.dato == referencia:
                    anterior = aux.anterior
                    siguiente = aux.siguiente
                    anterior.siguiente = siguiente
                    siguiente.anterior = anterior
                aux = aux.siguiente
                if aux == self.primero:
                    break
    
    #Método para mostrar los elementos de la lista 
    def recorrer_inicio(self):
        aux = self.primero
        while aux:
            print(aux.dato)
            aux = aux.siguiente
            if aux == self.primero:
                break

    #Método para mostrar los elementos de la lista en sentido inverso
    def recorrer_final(self):
        aux = self.ultimo
        while aux:
            print(aux.dato)
            aux = aux.anterior
            if aux == self.ultimo:
                break
```

* agregar_inicio(dato): Agrega un nuevo nodo al inicio de la lista. Si la lista está vacía, crea un nuevo nodo y lo establece como el primer y último nodo. En caso contrario, crea un nuevo nodo, establece el enlace siguiente al primer nodo actual, actualiza el enlace anterior del primer nodo y establece el nuevo nodo como el primer nodo.

* agregar_final(dato): Agrega un nuevo nodo al final de la lista. Si la lista está vacía, crea un nuevo nodo y lo establece como el primer y último nodo. En caso contrario, crea un nuevo nodo, establece el enlace siguiente del último nodo actual al nuevo nodo, actualiza el enlace anterior del nuevo nodo al último nodo y establece el nuevo nodo como el último nodo.

* eliminar_inicio(): Elimina el primer nodo de la lista. Si la lista está vacía, muestra un mensaje de error. Si el primer nodo es el único nodo en la lista, establece el primer y último nodo como None. En caso contrario, actualiza el primer nodo al siguiente nodo y actualiza los enlaces anterior y siguiente de los nodos correspondientes.

* eliminar_final(): Elimina el último nodo de la lista. Si la lista está vacía, muestra un mensaje de error. Si el último nodo es el único nodo en la lista, establece el primer y último nodo como None. En caso contrario, actualiza el último nodo al nodo anterior y actualiza los enlaces anterior y siguiente de los nodos correspondientes.

* agregar(dato, referencia): Agrega un nuevo nodo con el dato dado después del nodo de referencia en la lista. Si la lista está vacía, muestra un mensaje de error. Busca el nodo de referencia en la lista y enlaza el nuevo nodo después de él, actualizando los enlaces anterior y siguiente de los nodos correspondientes.

* eliminar(referencia): Elimina el nodo con el dato de referencia de la lista. Si la lista está vacía, muestra un mensaje de error. Busca el nodo de referencia en la lista y actualiza los enlaces anterior y siguiente de los nodos correspondientes para eliminar el nodo.

* recorrer_inicio(): Recorre la lista desde el primer nodo al último nodo, mostrando los datos de cada nodo en orden.

* recorrer_final(): Recorre la lista desde el último nodo al primer nodo, mostrando los datos de cada nodo en orden inverso.
